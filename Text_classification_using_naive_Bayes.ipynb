```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
df=pd.read_csv("/content/text_classification.csv")
```

```python
df
```

```python
df['label_num']=df.label.map({'pos':1,'neg':0})
df
```

```python
x=df.message
y=df.label_num
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
cv=CountVectorizer()
xtrain=cv.fit_transform(x_train)
xtest=cv.transform(x_test)
```

```python
mnb=MultinomialNB()
mnb.fit(xtrain,y_train)
y_pred=mnb.predict(xtest)
```

```python
print("Accuracy:",metrics.accuracy_score(y_pred,y_test))
print("Precision score:",metrics.precision_score(y_pred,y_test))
print("Recall score:",metrics.recall_score(y_pred,y_test))
```

```python
print("Confusion matrix:",confusion_matrix(y_pred,y_test))
```
